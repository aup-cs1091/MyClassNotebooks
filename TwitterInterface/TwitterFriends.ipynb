# This code retrieves the friends of a twitter account

Note that the imported file twurl.py is a py file and imports the py files oauth.py and hidden.py

In order to run this code you need to get authorization codes from twitter and place them into the hidden.py file
You can find simple instructions on how to get the twitter codes here: https://iag.me/socialmedia/how-to-create-a-twitter-app-in-8-easy-steps/ 

```python
import urllib.request, urllib.parse, urllib.error
import json
import ssl
```

```python
import twurl 
```

```python
# https://apps.twitter.com/
# Create App and get the four strings, put them in hidden.py

TWITTER_URL = 'https://api.twitter.com/1.1/friends/list.json'

# Ignore SSL certificate errors
ctx = ssl.create_default_context()
ctx.check_hostname = False
ctx.verify_mode = ssl.CERT_NONE

# this can be set to the desired number of friends
# numbOfFriends = 50

while True:
    # ask which account
    print('')
    acct = input('Enter Twitter Account (to re-run class example enter realDonaldTrump):')
    
    if (len(acct) < 1): break #if the user just enters return-carriage, stop
    print('')
    numbOfFriends = int(input('Enter how many friends you want to see (to re-run class example enter 50):'))
    url = twurl.augment(TWITTER_URL,
                        {'screen_name': acct, 'count': numbOfFriends})
    print('Retrieving', url)
    connection = urllib.request.urlopen(url, context=ctx)
    data = connection.read().decode()

    js = json.loads(data)
    print(json.dumps(js, indent=2))
    #save all data to file
    myfile = open('TwitterDump', 'w', encoding='utf-8')
    #json.dump(js, myfile) - use this if you want to read it back later in json format
    myfile.write(json.dumps(js, indent=2))
    myfile.close()

    headers = dict(connection.getheaders())
    print('Remaining runs', headers['x-rate-limit-remaining'])
    print('******** getting first ',str(numbOfFriends),' friends of ',acct,' *************')
    print('screen_name', '\t', 'name', '\t','followers_count','\t','location','\t','status')

    #open a file to add all data extracted from this account using utf-8 encoding for the text
    f = open('dataFromTwitter', 'a', encoding='utf-8')
    f.write("screen_name"+"\t"+"name"+"\t"+"followers_count"+"\t"+"location"+"\t"+"status"+"\n")

    for u in js['users']:
        if 'status' not in u:
            s='   * No status found'
        else:
            s = u['status']['text']
        try:
            print(u['screen_name'], '\t', u['name'], '\t',u['followers_count'],'\t',u['location'],'\t',s[:150])
        except:
            print("Unexpected error, something could not be printed")
        #write to file converting the number to string 
        f.write(u['screen_name']+"\t"+u['name']+"\t"+str(u['followers_count'])+"\t"+u['location']+"\t"+s[:150]+"\n")

    f.close()
```
